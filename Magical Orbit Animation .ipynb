## **Magical Orbit Animation (Static)**

```python
import cv2
import numpy as np
import math
import random
import time

# --- Window setup ---
width, height = 800, 800
center = (width // 2, height // 2)
cv2.namedWindow("Magical Orbit", cv2.WINDOW_FREERATIO)

# --- Create particles ---
num_particles = 100
particles = []
for i in range(num_particles):
    angle = random.uniform(0, 2 * math.pi)
    radius = random.uniform(80, 300)
    speed = random.uniform(0.01, 0.05)
    color = (
        random.randint(100, 255),
        random.randint(100, 255),
        random.randint(100, 255),
    )
    size = random.randint(2, 6)
    particles.append([angle, radius, speed, color, size])

# --- Animation loop ---
start_time = time.time()
duration = 60  # seconds

# for smooth trails
trail = np.zeros((height, width, 3), dtype=np.uint8)

while True:
    frame = (trail * 0.92).astype(np.uint8)  # fade old trails slightly

    # glowing center
    cv2.circle(frame, center, 10, (255, 255, 255), -1)
    cv2.circle(frame, center, 25, (150, 150, 255), 2)

    for p in particles:
        angle, radius, speed, color, size = p
        x = int(center[0] + math.cos(angle) * radius)
        y = int(center[1] + math.sin(angle) * radius)

        # draw particle
        cv2.circle(frame, (x, y), size, color, -1)

        # update angle for orbiting motion
        p[0] += speed

        # small random twinkle effect
        if random.random() < 0.02:
            p[3] = (
                random.randint(100, 255),
                random.randint(100, 255),
                random.randint(100, 255),
            )

    # text overlay
    cv2.putText(frame, "Magical Orbit Animation",
                (130, 50), cv2.FONT_HERSHEY_TRIPLEX, 0.8, (255, 255, 255), 2)

    cv2.imshow("Magical Orbit", frame)
    trail = frame.copy()

    # break conditions
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
    if time.time() - start_time > duration:
        break

cv2.destroyAllWindows()
```
